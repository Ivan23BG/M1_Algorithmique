```python
# Addition de deux polynomes
def polynomial_sum(f, g):
    """Adds two polynomials defined as a list and returns the result"""
    # Get the different list lengths needed later on
    n_f = len(f)
    n_g = len(g)
    n = max(n_f, n_g)

    # Initialise the output list
    output = [0] * n
    
    # Add f
    for i in range(n_f):
        output[i] += f[i]
    
    # Add g
    for i in range(n_g):
        output[i] += g[i]
    
    # Remove extra zeroes
    while output[n - 1] == 0:
        n -= 1
        if n == -1:
            return [0]

    return output[:n]
```

```python
# Test polynomial sum
f = [2, 2, -8]
g = [1/2, 7, 8]

h = polynomial_sum(f, g)
print(h)
```

```python
def polynomial_prod(f, g):
    """Multiplies two polynomials defined as a list and returns the result"""
    # Get the different list lengths needed later on
    n_f = len(f)
    n_g = len(g)
    n = max(n_f, n_g)

    # Swap the polynomials to have them in the right order (for convenience)
    if n_f > n_g:
        t = g
        g = f
        f = t
        
        t = n_g
        n_g = n_f
        n_f = t
    
    output = [0] * (n_f + n_g - 1)
    for i in range(n_f):
        output[i] += f[i] * g[0]
    for i in range(n_f, n_f + n_g - 1):
        output[i] += f[i - n_f + 1] * g[n_g - 1]

    output[n_f - 1] += f[0] * g[n_g - 1]

    for i in range(0, n_f):
        for j in range(1, n_g - 1):
            output[i + j] += f[i] * g[j]
    
    return output
```

```python
# Test the polynomial_prod
g = [2, 2, -8]
f = [1/2, 7, 8]

h = polynomial_prod(f, g)
print(h)
```
